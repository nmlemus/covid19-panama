```python
from pymc3 import Model, Normal, HalfNormal, Uniform

import warnings

import arviz as az
import matplotlib.pyplot as plt
import numpy as np
```

```python
def covid_prob(
        sensitivity=0.99,
        specificity=0.99,
        prevelance=0.01,
        verbose=True):
    """
    Computes the posterior using Bayes' rule
    """
    
    no_specificity = 1 - specificity
    no_covid = 1 - prevelance
    
    num = sensitivity*prevelance
    den = sensitivity*prevelance + no_specificity*no_covid
    
    p_covid_pos = num/den
    
    return p_covid_pos
```

```python
covid_prob(sensitivity=0.95, specificity=0.95, prevelance=0.34)
```

```python
ps = []
x = []
for pre in range(0,50):
    x.append(pre)
    p = covid_prob(sensitivity=0.95,specificity=0.95,prevelance=pre/100,verbose=False)
    ps.append(p)

plt.figure(figsize=(18,8))
#plt.title("Probabilidad de tener covid19, si el test fue po",fontsize=15)
plt.plot(x, ps,color='k',marker='o',markersize=4)
plt.grid(True)
plt.xlabel("Prevelancia (%)",fontsize=14)
plt.ylabel("Probabilidad de tener covid-19",fontsize=14)
plt.yticks(fontsize=12)
plt.show()
```

```python
ps = []
for sen in range(85, 99):
    sen = sen/100
    for spec in range(85, 99):
        spec = spec/100
        for pre in range(1,80):
            p = covid_prob(sensitivity=sen,specificity=spec,prevelance=pre/100,verbose=False)
            ps.append(p)
```
